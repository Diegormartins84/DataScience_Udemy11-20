```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
#pip install seaborn (executar no Anaconda Prompt)
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('dados.csv', sep=';')
base.shape
```

```python
base.head()
```

```python
srn.histplot(base.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(base.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
agrupado = base.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='PIB', color='gray').set(title='PIB por Município')
```

```python
agrupado2 = base.sort_values('VALOREMPENHO').head(10)
agrupado2 = agrupado2.iloc[:,[1,3]]
agrupado2
agrupado2.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color='gray').set(title='Empenho por Município')
```
